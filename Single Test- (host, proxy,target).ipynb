# <font color= #00008b> Single test </font>

This notebook contain single iperf3 test between one host to one target via 1 proxy.  
this test do the following:  
- Create host, proxy and target k8s clusters       
- Deploy iperf3 pods on each target and host cluster  
- Deploy forwarding and haproxy pods on proxy cluster  
Run 3 types of iperf3 test between host to target:  
  - Direct connection  
  - Using haproxy pod(tcp splitting)   
  - Using forwarding pod  
  
Before running this test please confirm:   
 1. The Required steup and installation are done see: [Pre Installation Requirements](./Jupyter_notebook/Pre_Installation_Requirements.ipynb)
 2. Login to all cloud providers(GCP,IBM,AWS): [Cloud Providers Login](./Jupyter_notebook/Cloud_Providers_Login.ipynb)

## Environment Setting (Cloud provider , locations)

Set hoat,target proxy locations and platforms:

```python
import os
HOST_ZONE   = "us-east1-b"    ; HOST_PLATFORM   = "gcp"; HOST_NAME   = "host-k8s"
TARGET_ZONE = "us-west1-a"    ; TARGET_PLATFORM = "gcp"; TARGET_NAME = "target-k8s"
PROXY_ZONE  = "us-central1-a" ; PROXY_PLATFORM  = "gcp"; PROXY_NAME  = "proxy-k8s"
```

## <font color= #00008b>Testing </font>

Run single test between host to target with the foloowing steps:

### Step 1: Cluster Creation

Creating hoat,target and proxy k8-s clusters: 

```python
import os
import sys
sys.path.insert(1, 'project_metadata/')
from tests.scripts.time_func import *
os.system("python3 ./steps/create_k8s_cluster.py -type host   -zone {} -platform {} -bg {}".format(HOST_ZONE,HOST_PLATFORM,True))
os.system("python3 ./steps/create_k8s_cluster.py -type proxy  -zone {} -platform {} -bg {}".format(PROXY_ZONE,PROXY_PLATFORM,True))
os.system("python3 ./steps/create_k8s_cluster.py -type target -zone {} -platform {} ".format(TARGET_ZONE,TARGET_PLATFORM))
```

### Step 2: Cluster readiness check 

Check all the clusters are deployed and ready:

```python
os.system("python3 ./steps/check_k8s_cluster_ready.py -type host   -zone {} -platform {} ".format(HOST_ZONE,HOST_PLATFORM))
os.system("python3 ./steps/check_k8s_cluster_ready.py -type proxy  -zone {} -platform {} ".format(PROXY_ZONE,PROXY_PLATFORM))
os.system("python3 ./steps/check_k8s_cluster_ready.py -type target -zone {} -platform {} ".format(TARGET_ZONE,TARGET_PLATFORM))
```

### Step 3:Clusters setting

setting the k8s cluster on the server according to his type:
- host/target - iperf3 deployment
- proxy       - forwarding/haproxy deployment


```python
#setting source, target, proxy clusters
os.system("python3 ./steps/set_k8s_cluster.py -type target -zone {} -platform {}".format(TARGET_ZONE,TARGET_PLATFORM))
os.system("python3 ./steps/set_k8s_cluster.py -type host   -zone {} -platform {}".format(HOST_ZONE,HOST_PLATFORM))
os.system("python3 ./steps/set_k8s_cluster.py -type proxy  -zone {} -platform {}".format(PROXY_ZONE,PROXY_PLATFORM))
```

### Step 4: Single Iperf3 Runs


4.1 get IP and Port of all the servers and results directory for the test oputputs(JSON file):

```python
from steps.aux_func.network_test_func import run_single_iperf_test,run_multiple_iperf_test,connect_to_cluster,get_folder_res
from project_metadata.meta_data_func import get_ip_and_port
from steps.PROJECT_PARAMS import METADATA_FILE

host_ip       , host_port       = get_ip_and_port(METADATA_FILE, HOST_NAME  , HOST_ZONE,   type_data="host")
target_ip     , target_port     = get_ip_and_port(METADATA_FILE, TARGET_NAME, TARGET_ZONE, type_data="target")
proxy_ip      , proxy_port      = get_ip_and_port(METADATA_FILE, PROXY_NAME , PROXY_ZONE,  type_data="proxy")
forwarding_ip , forwarding_port = get_ip_and_port(METADATA_FILE, PROXY_NAME , PROXY_ZONE,  type_data="forwarding")

res_dir = get_folder_res(HOST_ZONE,HOST_PLATFORM, TARGET_ZONE, TARGET_PLATFORM, PROXY_ZONE, PROXY_PLATFORM,folder_type="proxy_folder")
print(res_dir)
```

4.2 Run measurements form proxy to host and target

```python
#Set NOF Tests
NOF_TESTS=10
```

```python
#Test on Proxy clusters
connect_to_cluster(PROXY_NAME,PROXY_ZONE,PROXY_PLATFORM)
```

```python
run_multiple_iperf_test(PROXY_NAME, TARGET_NAME, target_ip, target_port, res_dir, type_data="direct",nof_tests=NOF_TESTS)
```

```python
run_multiple_iperf_test(PROXY_NAME, HOST_NAME,   host_ip,   host_port,   res_dir, type_data="direct",nof_tests=NOF_TESTS)
```

4.2 Run measurements form host to target

```python
#Test on hosts clusters
connect_to_cluster(HOST_NAME,HOST_ZONE,HOST_PLATFORM)
```

```python
#Direct test between host to target
run_multiple_iperf_test(HOST_NAME, TARGET_NAME, target_ip,    target_port,     res_dir, type_data="direct", nof_tests=NOF_TESTS)
```

```python
#Test between host to target with splitting proxy
run_multiple_iperf_test(HOST_NAME, PROXY_NAME,  proxy_ip,     proxy_port,      res_dir, type_data="proxy", nof_tests=NOF_TESTS)
```

```python
#test between host to target with forwarding proxy
run_multiple_iperf_test(HOST_NAME, PROXY_NAME,  forwarding_ip, forwarding_port, res_dir, type_data="forwarding", nof_tests=NOF_TESTS)
```
